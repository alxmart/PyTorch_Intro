```python
import pandas as pd
import torch
import numpy as np
```

```python
# define the apartment's rent

rent = 2550

# convert to an integer tensor

rent_tensor = torch.tensor(
    rent,
    dtype=torch.int)
```

```python
print(rent_tensor)
```

```python
print(rent_tensor.dtype)
```

```python
# create a numpy array with the rent in US dollars,
# size in square feet, and age in years

apt_array = np.array([2550, 750, 3.5])

# convert to a tensor of floats

apt_tensor = torch.tensor(
    apt_array,
    dtype=torch.float)
```

```python
print(apt_tensor)
```

```python
print(apt_tensor.dtype)
```
